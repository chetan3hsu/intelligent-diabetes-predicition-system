# Diabetes Prediciton Using Logistic Regression

## Goal:
### To build out a Logistic Regression model using SGPGI Hospital data for diabetes predicition.

## Importing Required Library

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
import pickle
```

## Loading Dataset


```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/patients_data.xlsx')
df.head()
```

## Data Preprocessing


### Removing Unnecessary Features

```python
df.drop(columns=['Treatment', 'Cholestrol2','CRP1', 'CRP2', 'CRP3', 'AgeGroups', 'haemoglobingroups' ])
```

### Data Encoding


```python
df_model=df.copy()
binary = {1:1, 2:0}
colms=['Gender',	'Obesity', 'Diabetes',	'Hypertension','Blood_pressure',	'Smoker',	'Cholestrol', 'Angina',]

```

### Data Splitting


```python
X=df_model.drop('Diabetes', axis=1)
y=df_model['Diabetes']
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=y, random_state=42)
```

### Class Balancing

```python
sm=SMOTE(random_state=42)
X_train_bal, y_train_bal=sm.fit_resample(X_train,y_train)
```

## Logistic Regression Model

#### Model Building

```python
lr_model=LogisticRegression(max_iter=1000, class_weight="balanced")
lr_model.fit(X_train_bal, y_train_bal)
```

### Pickle Model

```python
pickle.dump(
    lr_model, open('logisticReg_model.pickle', 'wb')
)
```
